```python
import pandas as pd
import matplotlib.pyplot as plt

# File paths
file1 = r'C:\Users\newbr\OneDrive\Desktop\B_model_87\classification_report_sfnoise_k.csv'
file2 = r'C:\Users\newbr\OneDrive\Desktop\B_model_87\classification_report_sf_k.csv'

# Read the CSV files
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

# Extract the f1-scores
f1_scores1 = df1['f1-score']
f1_scores2 = df2['f1-score']

# Calculate the differences
f1_diff = f1_scores1 - f1_scores2
mean_diff = f1_scores1.mean() - f1_scores2.mean()
median_diff = f1_scores1.median() - f1_scores2.median()
std_diff = f1_scores1.std() - f1_scores2.std()

# Plot the f1-scores
plt.figure(figsize=(10, 6))
plt.plot(f1_scores1, label='SF Noise', marker='o')
plt.plot(f1_scores2, label='SF', marker='o')
plt.xlabel('Class')
plt.ylabel('F1-Score')
plt.title('Comparison of F1-Scores')
plt.legend()
plt.grid(True)
plt.show()

# Plot the differences
plt.figure(figsize=(10, 6))
plt.plot(f1_diff, label='Difference', linestyle='--', color='red', marker='x')
plt.xlabel('Class')
plt.ylabel('Difference in F1-Score')
plt.title('Difference in F1-Scores')
plt.legend()
plt.grid(True)
plt.show()

# Plot mean difference
plt.figure(figsize=(10, 6))
plt.bar(['Mean Difference'], [mean_diff], color='blue')
plt.ylabel('Mean Difference in F1-Score')
plt.title('Mean Difference in F1-Scores')
plt.grid(True)
plt.show()

# Plot median difference
plt.figure(figsize=(10, 6))
plt.bar(['Median Difference'], [median_diff], color='green')
plt.ylabel('Median Difference in F1-Score')
plt.title('Median Difference in F1-Scores')
plt.grid(True)
plt.show()

# Plot standard deviation difference
plt.figure(figsize=(10, 6))
plt.bar(['Standard Deviation Difference'], [std_diff], color='purple')
plt.ylabel('Standard Deviation Difference in F1-Score')
plt.title('Standard Deviation Difference in F1-Scores')
plt.grid(True)
plt.show()
```
